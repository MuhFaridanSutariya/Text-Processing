```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
text = "Halo, nama saya faridan. Saya beli batagor dengan harga Rp 10.000 hari ini. Kamu mau?"
```

# Preprocessing

## Normalization

```python
text.lower()
```

## Tokenization

```python
# memisahkan setiap sentences (paragraf)
# itu lah sebab nya kita mendownload punkt yang akan memisahkan setiap paragraf
sent_tokenize(text)
```

```python
# memisahkan setiap kata
tokens = word_tokenize(text)
tokens
```

## Punctuation Removal / Alphanumeric Cleansing
#### membuat tanda baca

```python
punctuation
```

```python
tokens = [token for token in tokens if token not in punctuation]
tokens
```

```python
tokens = [token for token in tokens if token.isalnum()]
tokens
```

## stop words removal
#### digunakan untuk mencari kata-kata yang paling sering muncul

```python
stopwords.words("indonesian")
```

```python
tokens = [token for token in tokens if token not in stopwords.words("indonesian")]
tokens
```
